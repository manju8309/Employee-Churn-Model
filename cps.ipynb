```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

n_samples = 1000

labels = np.random.choice([0, 1], size=n_samples, p=[0.7, 0.3])

sensor_data = np.random.normal(0, 1, (n_samples, 10))

accuracy = {
    "SVM": 91.4,
        "RF": 93.8,
            "CNN": 95.2,
                "LSTM": 96.1,
                    "CNN-LSTM": 98.6
                    }
plt.figure(figsize=(8, 5))
plt.plot(sensor_data[:100])
plt.title("CPS Monitoring Interface")
plt.xlabel("Time")
plt.ylabel("Sensor Values")
plt.grid()
plt.show()

plt.figure(figsize=(8, 5))
plt.bar(accuracy.keys(), accuracy.values())
plt.title("Detection Results of Models")
plt.ylabel("Accuracy (%)")
plt.ylim(85, 100)
plt.grid(axis='y')
plt.show()

normal_scores = np.random.normal(0.2, 0.05, 500)
attack_scores = np.random.normal(0.7, 0.1, 500)

plt.figure(figsize=(8, 5))
sns.histplot(normal_scores, label='Normal', kde=True)
sns.histplot(attack_scores, label='Attack', kde=True)
plt.title("Anomaly Score Distribution")
plt.legend()
plt.show()

prob_normal = np.random.uniform(0.6, 1.0, 200)
prob_attack = np.random.uniform(0.0, 0.4, 200)

plt.figure(figsize=(8, 5))
plt.plot(prob_normal, label="Normal Probabilities")
plt.plot(prob_attack, label="Attack Probabilities")
plt.title("Classification Probability Graph")
plt.legend()
plt.show()

recon_normal = np.random.normal(0.1, 0.02, 300)

plt.figure(figsize=(8, 5))
plt.plot(recon_normal)
plt.axhline(y=0.2, linestyle='--', label='Threshold')
plt.title("Reconstruction Error (Normal)")
plt.legend()
plt.show()

recon_attack = np.random.normal(0.35, 0.08, 300)

plt.figure(figsize=(8, 5))
plt.plot(recon_attack)
plt.axhline(y=0.2, linestyle='--', label='Threshold')
plt.title("Reconstruction Error (Anomalous)")
plt.legend()
plt.show()
```
